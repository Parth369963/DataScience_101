# Date Time Manipulation


```python
date_range = pd.date_range('4/12/2021', periods=12,freq='M')   # freq='M' means time frequency = months'M', years''Y

dates = pd.DataFrame(date_range,columns=['Date'])
```

```python
dates
```

```python
dates['Day'] = dates['Date'].dt.day
dates['Month'] = dates['Date'].dt.month
dates['Year'] = dates['Date'].dt.year

dates
```

# Example to convert object date into proper date format

```python
df = pd.DataFrame({'Date_Time':['Jan 1 12:00 am','Jan 2 12:00 am']})

df
```

```python
df['Date'] = pd.to_datetime(df['Date_Time'],format='%b %d %I:%M %p')  # This formate and how to change is provided in below link
df
```

## https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior
