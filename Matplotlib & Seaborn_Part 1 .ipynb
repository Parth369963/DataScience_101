```python
sal = [15,22,24,25,30,34,23,43,13,20,25,52]
sal.sort()
sal
```

Please see below example,

[15,22,24,25,30,34,23,43,13,20,25,52]

n = total elements in the above example(12 items)

25% =(25/100)*(n+1)=3.25(data at 3.25 position is the 25% percentile. let's say around 20.50)

50% =(50/100)*(n+1)=6.5 (data at this position)

75% =(75/100)*(n+1)=9.75 (data at this position) 

# Outliers

```python
ser = pd.Series(sal)
```

```python
ser
```

```python
#1st quantile
q1 = ser.quantile(0.25)
print ('Q1', q1)

#3rd quantile
q3 = ser.quantile(0.75)

print ('Q3', q3)

# Find the IQR (inter quantile range) to identify outliers

iqr = q3-q1

print('IQR',iqr)
```

## Outlier detection formula

higher side ==> Q3 + (1.5 * IQR)

lower side ==> Q1 - (1.5 * IQR)



- Maximum (Max)
- Qauntile 3 (Q3)
- Qauntile 2 (Q2)
- Qauntile 1 (Q1)
- Minimum (Min)

```python
upside = q3 + (1.5 * iqr)
print (upside)
```

```python
lower_side = q1 - (1.5 * iqr)
print (lower_side)
```

# Pie Chart

```python
rupees_monthly = [2000,2530,900,15000,1500]
items = ['Tution_fee','Mobile_bill','Electricity_bill','Rent','Misc']

#-----------------Plotting Data---------------------

plt.pie(x=rupees_monthly, labels=items)
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.pie(x=rupees_monthly, labels=items,explode=[0.4,0,0.3,0,0],autopct = '%.2f', rotatelabels=True)
plt.show()

# ['Tution_fee','Mobile_bill','Electricity_bill','Rent','Misc'] = [0.4,0,0.3,0,0]
```

Here, explode=[0.3,0,0,0] which means you are highlighting and showing a specific data from the pie chart were [0,0,0,0] are the indexes.

- If you mention .3 in index e.g. [0.3,0,0,0,0] where 0.3 is showing exploding the 'Tution_fee' in piechart

- If you increase the decimal .4,.5, or .6 it is used to how much pie we want to explode

['Tution_fee','Mobile_bill','Electricity_bill','Rent','Misc'] = [0,0,0,0,0]


Here, autopct = '%.2f' which means percentage and .2 is 2 decimal if you write '%.3f' it will show percentage with 3 decimals

Here, rotatelabels=True is used to make the labels bit inclined (visualization purpose)

# relplot()

- This function provides access to several different axes-level functions that show the relationship between two variables with semantic mappings of subsets. 

```python
tips = sns.load_dataset('tips') # Loading a preloaded free dataset

tips.tail()
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips)
plt.show()

tips['smoker'].value_counts()
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker', style = 'time')
plt.show()

# This way we can understand what time they visited, were they smoker or not, were they came for dinner or lunch
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, size = 'day')
plt.show()

# Checking who visited on which weekdays
```

# scatterplot()

- The scatterplot is a plot with many data points. It is one of the many plots seaborn can create. Seaborn can create this plot with the scatterplot() method. The data points are passed with the parameter data.

```python
sns.scatterplot(x = 'total_bill', y = 'tip', data = tips, size = 'size', hue = 'sex')

plt.show()
```

# lineplot()

- Seaborn Line Plots depict the relationship between continuous as well as categorical values in a continuous data point format.

```python
sns.lineplot(x = 'total_bill', y = 'tip', data = tips)
plt.show()

# Lineplot will connect all the dots from scatter plot
```
