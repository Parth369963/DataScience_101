### LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lab_enc = LabelEncoder() # This will assign/place data into numerical number as 0,1,2,3
```

```python
df2 = lab_enc.fit_transform(df['City'])  
pd.Series(df2)

# example ('Bengaluru' = 0, 'Delhi' = 1, 'Hyderabad' = 2, 'Bengaluru' = 3, 'Hyderabad' = 4, 'Bengaluru' = 5)

# So whenever there is Bengaluru in DataFrame it will replace it with 0, when Hyderabad it will replace it with 2
```

```python
df['City'] = df2   # Replacing 'City' with df2 data 
df
```

### OneHotEncoder

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer             # Importing simple imputer
from sklearn.compose import make_column_transformer  # used to join different sets of columns after processing
```

```python
ohe = OneHotEncoder()
si = SimpleImputer()
```

```python
import pandas as pd

df = pd.DataFrame({'Salary' : [25000,48000,71000,85000,90000,55000], 
                   'City' : ['Bengaluru','Delhi','Hyderabad','Bengaluru','Hyderabad','Bengaluru'], 
                   'Gender' : ['Male','Female','Female','Female','Male','Male'], 
                   'Exp' : [1,3,5,6,9,None]})

df
```

```python
ct = make_column_transformer((ohe,['City','Gender']),
                             (si, ['Exp']),              # si, ['Exp'] will take mean value from 'Exp' & will replace it.
                             remainder = 'passthrough')  # 'passthrough' to keep all other columns.
```

```python
encoded = pd.DataFrame(ct.fit_transform(df))
encoded
```

```python
# Rename the columns as per our choice.

encoded = pd.DataFrame(ct.fit_transform(df),columns=['City_Bengaluru', 'City_Delhi','City_Hyd','Gender_Male','Gender_Female','Exp','Salary'])
```

```python
encoded
```

### get_dummies
- One hot Encoding and get_dummies almost equal. Major difference is if you want to reduce (drop_first = True) the column size of the dataset you can use get_dummies.
- OHE does not add variable names to your data frame, but get_dummies add variable names.
- Sometimes having more columns might overfit the model

```python
df1 = pd.get_dummies(df[['City','Gender']])
df1
```

```python
df1 = pd.get_dummies(df[['City','Gender']],drop_first=True)
df1
```

### Knn Imputer

```python
df = pd.DataFrame({'Salary':[25000,48000,71000,85000,90000,49000],
                   'City':['Bengaluru','Delhi','Hyderabad','Bengaluru','Hyderabad','Bengaluru'],
                   'Gender':['Male','Female','Female','Female','Male','Male'],
                   'Exp':[1,3,5,6,9,None]})

df
```

```python
# Knn imputer will try to find the relation with other columns and impute the data according the relation
# In this case Age NaN is depending on the similarity with Fare columns

from sklearn.impute import KNNImputer
```

```python
Knnimp = KNNImputer(n_neighbors=2)
Knn_imp = pd.DataFrame(Knnimp.fit_transform(df[['Salary','Exp']])) 

Knn_imp
```

# Iterative Imputer

**This method treat other columns (which doesnot have nulls as feature and train on them and treat Null column as label. Fianlly it will predict the NaN data and impute. Its just like regression problem. Here Null column is label.**

```python
# Before using Iterative Imputer, We need to enable it using below code

from sklearn.experimental import enable_iterative_imputer

# Import Iterative Imputer

from sklearn.impute import IterativeImputer
```

```python
df = pd.DataFrame({'Salary':[25000,48000,71000,85000,90000,55000],
                   'City':['Bengaluru','Delhi','Hyderabad','Bengaluru','Hyderabad','Bengaluru'],
                   'Gender':['Male','Female','Female','Female','Male','Male'],
                   'Exp':[1,3,5,6,9,None]})

df
```

```python
iter_impute = IterativeImputer()
ite_imp = pd.DataFrame(iter_impute.fit_transform(df[['Salary','Exp']]),columns=['Salary','Exp'])
ite_imp
```
